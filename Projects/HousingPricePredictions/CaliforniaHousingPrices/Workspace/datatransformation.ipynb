```python

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import numpy as np
```

```python
path = 'E:/ML & Data/Projects/CaliforniaHousingPrices/Data/housing.csv'
df_main = pd.read_csv(path)
df_main
```

```python

df_main_cleaned = df_main[df_main.median_house_value < 500000]
df_main_cleaned = df_main_cleaned.reset_index(drop=True)
df_main_cleaned
```

```python

df_inputs = df_main_cleaned.drop('median_house_value',axis=1)
df_labels = df_main_cleaned['median_house_value']
df_inputs
```

```python
encoder = OneHotEncoder()
df_cat = df_inputs[['ocean_proximity']]
df_cat_encoded = encoder.fit_transform(df_cat)
df_cat_encoded = df_cat_encoded.toarray()
columns = encoder.get_feature_names_out()
final_columns = [name.split('_')[2] for name in columns]
df_cat_num = pd.DataFrame(df_cat_encoded,columns=final_columns)
df_cat_num_fixed = df_cat_num.drop("NEAR OCEAN",axis=1)
df_cat_num_final = pd.concat([df_inputs.drop("ocean_proximity",axis=1),df_cat_num_fixed],axis=1)
df_cat_num_final.head(3)
```

```python
inputer = SimpleImputer(strategy='median')
inputed_data = inputer.fit_transform(df_cat_num_final)
inputed_data = pd.DataFrame(inputed_data,columns=df_cat_num_final.columns)
inputed_data.head()
inputed_data.info()
```

```python
df_scalling = inputed_data.drop(['<1H OCEAN','INLAND','ISLAND','NEAR BAY'],axis=1)
scaller = MinMaxScaler()
scaller_fit = scaller.fit_transform(df_scalling)
df_scalled = pd.DataFrame(scaller_fit,columns=df_scalling.columns)
```

```python
df_final = pd.concat([df_scalled,df_cat_num_fixed],axis=1)
df_final
```

```python
df_final.to_csv("Scalled_Data",index=False)
```
