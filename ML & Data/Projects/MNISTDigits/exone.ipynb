```python

def modify_data(image_data,direction):
    image_data = image_data.reshape(28,28)
    if direction == 'up':
        for i in range(len(image_data)-1):
            for j in range(len(image_data[i])):
                image_data[i][j] = image_data[i+1][j]
    
    elif direction == 'down':
        temp_down = image_data[::-1]
        for i in range(len(temp_down)-1):
            for j in range(len(temp_down[i])):
                temp_down[i][j] = temp_down[i+1][j]
        image_data = temp_down[::-1]
    
    elif direction == 'left':
        for i in range(len(image_data)):
            for j in range(len(image_data[i])-1):
                image_data[i][j] = image_data[i][j+1]
    
    elif direction == 'right':
        for i in range(len(image_data)):
            for j in range(len(image_data[i])-1,-1,-1):
                image_data[i][j] = image_data[i][j-1]
    return image_data
```

```python
operatonal_data_up = data.iloc[:].to_numpy().copy()
operatonal_data_down = data.iloc[:].to_numpy().copy()
operatonal_data_right = data.iloc[:].to_numpy().copy()
operatonal_data_left = data.iloc[:].to_numpy().copy()
```

```python
new_up = []
new_down = []
new_right = []
new_left = []

for sample in operatonal_data_up:
    new_up.append(modify_data(sample,'up').reshape(784))
    new_down.append(modify_data(sample,'down').reshape(784))
    new_right.append(modify_data(sample,'right').reshape(784))
    new_left.append(modify_data(sample,'left').reshape(784))
    
```
