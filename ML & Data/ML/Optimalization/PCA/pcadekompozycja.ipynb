```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
```

```python
df
```

```python
df_inputs = df.drop(['HeartDisease'],axis=1).copy()
df_inputs.head(3)
```

```python
df_target = df['HeartDisease'].copy()
```

```python
df_non_numeric = df[['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']]
df_numeric = df_inputs.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)
df_scaled = pd.DataFrame(df_scaled,columns=df_numeric.columns)
```

```python
df_scaled
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
encoded = ohe.fit_transform(df_non_numeric)
encoded = encoded.toarray()
columns = ohe.get_feature_names_out()
final_columns = [name.split('_')[1] for name in columns]
encoded = pd.DataFrame(encoded,columns = final_columns)
```

```python
final_data = pd.concat([df_scaled,encoded],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(final_data,df_target,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(df_scaled)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pca,df_target,test_size = 0.2)
```

```python
model_pca = LogisticRegression()
model_pca.fit(X_train,y_train)
```

```python
model_pca.score(X_test,y_test)
```
