```python
import pandas as pd
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head(5)
```

```python
df.nunique()
```

```python
df.shape
```

```python
df.info()
```

```python
dataset = df[['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']].copy()
dataset
```

```python
dataset.isna().sum()
```

```python
dataset[['Bedroom2','Bathroom','Car','Distance']] = dataset[['Bedroom2','Bathroom','Car','Distance']].fillna(0)
dataset.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
to_imput = dataset[["Landsize",'BuildingArea']]
imputed = imputer.fit_transform(to_imput)
imputed = pd.DataFrame(imputed,columns=['Landsize','BuildingArea'])
imputed.head(5)
```

```python
dataset[['Landsize','BuildingArea']] = imputed
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset = dataset.reset_index()
```

```python
dataset = dataset.drop('index',axis=1)
```

```python
dataset
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
to_encode = dataset[['Suburb','SellerG','Regionname','CouncilArea','Type','Method']]
encoded = encoder.fit_transform(to_encode)
encoded = encoded.toarray()
columns = encoder.get_feature_names_out()
final_columns = [name.split('_')[1] for name in columns]
encoded = pd.DataFrame(encoded,columns = final_columns)
encoded.head(5)
```

```python
final_dataset = dataset.copy()
final_dataset = final_dataset.drop(['Suburb','SellerG','Regionname','CouncilArea','Type','Method'],axis=1)
final_dataset = pd.concat([final_dataset,encoded],axis=1)
final_dataset.head(5)
```

```python
df_input = final_dataset.drop('Price',axis=1).copy()
df_target = final_dataset['Price']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scalled_input= scaler.fit_transform(df_input)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(scalled_input,df_target,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```
