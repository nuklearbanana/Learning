```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import datasets
wine = datasets.load_wine()
dir(wine)
```

```python
wine.target_names
```

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df
```

```python
scaller = MinMaxScaler()
df_scalled = pd.DataFrame(scaller.fit_transform(df),columns=df.columns)
df_scalled.head()
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.pipeline import Pipeline
m_model = MultinomialNB()
g_model = GaussianNB()
```

```python
mp = Pipeline([
    ('Scaler',MinMaxScaler()),
    ('Model',MultinomialNB())
])

gp = Pipeline([
    ('Scaler',MinMaxScaler()),
    ('Model',GaussianNB())
])
```

```python
target = wine.target
X_train,X_test,y_train,y_test = train_test_split(df,target,test_size=0.3,random_state=100)
```

```python
mp.fit(X_train,y_train)
gp.fit(X_train,y_train)
```

```python
mp.score(X_test,y_test)
```

```python
gp.score(X_test,y_test)
```

Gauussowskie jest lepsze

```python
from sklearn.model_selection import cross_val_score
cross_val_score(mp,df_scalled,target)
```

```python
cross_val_score(gp,df_scalled,target)
```
